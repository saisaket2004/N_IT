```python
import pandas as pd

df1 = pd.DataFrame({
    'destination': ['Home', 'Work', 'No Urgent Place', 'Mall'],
    'weather': ['Sunny', 'Rainy', 'Sunny', 'Snowy'],
    'temperature': [55, 60, 80, 30]
})
```

```python
# combining two datasets and calling distinct values in a column
#SELECT DISTINCT destination FROM(SELECT * FROM dataset_1 UNION SELECT * FROM table_to_union)
pd.concat([df, df1])['destination'].drop_duplicates()
```

```python
df2 = pd.DataFrame({
    'time': ['7AM', '10AM', '2PM', '6PM', '10PM'],
    'part_of_day': ['Morning', 'Morning', 'Afternoon', 'Evening', 'Night']
})
```

```python
# combining two datasets based on a common column and calling specific columns in the combined dataset
#SELECT a.destination,a.time,b.part_of_day FROM dataset_1 a INNER JOIN table_to_join b ON a.time=b.time 
pd.merge(df, df2[['time', 'part_of_day']], on='time', how='inner')[['Destination', 'time', 'part_of_day']] 
```

```python
# combining two datasets based on a common column and calling specific columns in the combined dataset with a filter condition
#SELECT destination ,passenger FROM(SELECT*FROM dataset_1 WHERE passenger = 'Alone')
df[df['passanger'] == 'Alone'][['Destination', 'passanger']]
```

```python
# filtering the dataset based on a condition with a wildcard character
#SELECT * FROM dataset_1 WHERE weather LIKE 'Sun%'
df[df['weather'].str.startswith('Sun')]
```

```python
# filtering the dataset based on a condition with a range of values
#SELECT DISTINCT temperature FROM dataset_1 WHERE temperature BETWEEN 29 AND 75
df[(df['temperature'] >= 29) & (df['temperature'] <= 75)]['temperature'].unique() 
```

```python
# filtering the dataset based on a condition with a list of values
#SELECT occupation FROM dataset_1 WHERE occupation IN('Sales & Related','Management')
df[df['occupation'].isin(['Sales & Related', 'Management'])][['occupation']]
```
