```python
#EDA is data cleaning technique to convert raw data to clean data and build ml models
# eda has 6 techniques
#1 variable identification
#2 univariate analysis
#3 bivariate analysis
#4 outlier treatment
#5 missing value treatment
#6 imputation technique| transform
```

```python
#1 variable identification
# variable identification is the process of identifying the type of variable in the dataset 
# variable can be of 3 types
#1 categorical variable
#2 numerical variable
#3 datetime variable

#independent variable ( x1, x2) 
#dependent variable (y) 
#relevant attribute 
#irrelevant attribute

#every time try to build the model with relevant attribute only 
#if user build the model by irrelevant attribute then overfitting problem or multicollinearity problem will occur 
#model behave wrong -- less accuracy and high error 
```

```python
#2- univariate analysis
# univariate analysis is the process of analyzing the distribution of a single variable
# univariate analysis is used to understand the distribution of the variable and to identify the outliers in the variable
# univariate analysis is used to identify the central tendency of the variable and to identify the spread of the variable
#plotting the graph using 1 variable
```

```python
#3- bivariate anlaysis
# bivariate analysis is the process of analyzing the relationship between two variables
# bivariate analysis is used to understand the relationship between two variables and to identify the outliers in the relationship between two variables
# plot the graph using 2 variable
# corelation -- co + relation 
# relation amoung 2 attribute is called corealation 
# range of corelation to -1 to 1
# +ve corr -  0- 1
#-ve corr -- -1 to 0
# 0 corr -- 0 | no corelation
# corelation measure the strength of the relationship between two variables
```

```python
#4- outlier treatment 
# outlier is a data point that is significantly different from other data points in the dataset
```

```python
#5- missing value treatment
# missing value is a value that is not available in the dataset
# if numerical values are missing then mean , median, mode strategy 
# if categorical values are missing then mode strategy
# what if all value mode is same then what next ????
# knn (k nearest neighbour) strategy
# knn is a machine learning algorithm that is used to impute the missing values in the dataset
```

```python
#6- imputation technique also called as trasformer 
# imputation technique is used to transform the data in the dataset
# imputation technique is used to transform the data in the dataset to make it suitable for machine
# convert categorical value to numberical is called impution technique or transformer 
# 1 - dummy variable | one hot encoder --  0 or 1 
# 2- labelencoder 
# nlp the prcess of covert text or sent to vector -- embedding concept ( text -- number) 
# .get_dummies -- function used for transformer transfromact categorical to nuber 
# one hot encoder 
# labelconder 
```

# Data Cleaning

```python
emp['Name']
```

```python
emp['Name'] = emp['Name'].str.replace(r'\W',"",regex = True) #non word char
```

```python
emp["Name"]
```

```python
emp['Domain'] = emp['Domain'].str.replace(r'\W',"",regex = True) #non word char
```

```python
emp["Domain"]
```

```python
emp['Age'] = emp['Age'].str.replace(r'\W',"",regex = True) 
# this is how the data is looking like after removing the non word char from the age column
```

```python
emp['Age']
```

```python
emp['Age'] = emp['Age'].str.extract(r'(\d+)') #r(r'(\\d+)')
# this is how the data is looking like after extracting the digits from the age column
```

```python
emp["Age"]
```

```python
emp['Location'] = emp['Location'].str.replace(r'\W',"",regex = True) #non word char
```

```python
emp["Location"]
```

```python
emp['Salary'] = emp['Salary'].str.replace(r'\W',"",regex = True) #non word char
```

```python
emp['Salary']
```

```python
emp['Exp'] = emp['Exp'].str.extract(r'(\d+)') #r(r'(\\d+)')
```

# Let Us Apply EDA Techniques

## Missing Value Treatment

```python
clean_data.isnull().sum()
```

```python
clean_data['Age']
```

```python
import numpy as np
```

```python
clean_data['Age'] = clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data["Age"])))
```

```python
clean_data['Age']
```

```python
clean_data["Exp"]
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data["Exp"])))
```

```python
clean_data["Exp"]
```

```python
clean_data['Location'] = clean_data['Location'].fillna(clean_data["Location"].mode()[0])
```

```python
clean_data
```

```python
clean_data.info()
```

```python
clean_data["Age"] = clean_data['Age'].astype(int)
# changing the data type of age column from object to int
```

```python
clean_data['Age']
```

```python
clean_data.info()
```

```python
clean_data["Salary"] = clean_data['Salary'].astype(int)
clean_data["Exp"] = clean_data['Exp'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data["Name"] = clean_data['Name'].astype('category')
clean_data["Domain"] = clean_data['Domain'].astype('category')
clean_data["Location"] = clean_data['Location'].astype('category')
```

```python
clean_data.info()
```

```python
clean_data.to_csv('clean_data.csv')
```
