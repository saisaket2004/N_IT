```python
# There are 17 built-in functions for set in python
# 1. add() - adds an element to the set
# 2. clear() - removes all elements from the set
# 3. copy() - returns a copy of the set
# 4. update() - updates the set with the union of itself and others
# 5. remove() - removes an element from the set. Raises KeyError if the element is not present
# 6. discard() - removes an element from the set if it is present
# 7. pop() - removes and returns an arbitrary element from the set. Raises KeyError if the set is empty
# 8. union() - returns the union of sets as a new set
# 9. intersection() - returns the intersection of sets as a new set
# 10. difference() - returns the difference of sets as a new set
# 11. symmetric_difference() - returns the symmetric difference of sets as a new set
# 12. intersection_update() - updates the set with the intersection of itself and others
# 13. difference_update() - updates the set with the difference of itself and others
# 14. symmetric_difference_update() - updates the set with the symmetric difference of itself and others
# 15. issubset() - returns True if the set is a subset of another set
# 16. issuperset() - returns True if the set is a superset of
# 17. isdisjoint() - returns True if the set has no elements in common with another set
```

```python
# Set Operations
```

```python
s7 = {1,2,3,4,5} 
s8 = {4,5,6,7,8} 
s9 = {8,9,10} 
```

```python
s7.union(s8)
# The union() method returns a new set that contains all the elements from both sets, without duplicates. In this case, the union of s7 and s8 will be {1, 2, 3, 4, 5, 6, 7, 8}.
```

```python
s7.union(s8, s9)
```

```python
s8 | s9 
# The | operator is used to perform the union of two sets. In this case, s8 | s9 will return a new set that contains all the unique elements from both sets, which is {4, 5, 6, 7, 8, 9, 10}.
```

```python
s8 | s9 | s7 
```

```python
print(s7)
print(s8)
print(s9)
```

```python
s7.intersection(s8)
# The intersection() method returns a new set that contains only the elements that are present in both sets. In this case, the intersection of s7 and s8 will be {4, 5} because those are the only elements that are common to both sets.
```

```python
s7.intersection(s9)
# The intersection of s7 and s9 will be an empty set because there are no common elements between the two sets. Therefore, the result will be set() or {} depending on how it is displayed.
```

```python
s8 & s9 
# The & operator is used to perform the intersection of two sets. In this case, s8 & s9 will return a new set that contains only the elements that are present in both sets, which is {8} because 8 is the only element that is common to both sets.
```

```python
s7 = {1,2,3,4,5} 
s8 = {4,5,6,7,8} 
s9 = {8,9,10} 
```

```python
s9.intersection_update(s8)
```

```python
s9
```

```python
s7.difference(s8) 
# The difference() method returns a new set that contains the elements that are present in the first set but not in the second set. In this case, s7.difference(s8) will return {1, 2, 3} because those are the elements that are in s7 but not in s8.
```

```python
s8.difference(s7)
```

```python
s8 - s9
# The - operator is used to perform the difference of two sets. In this case, s8 - s9 will return a new set that contains the elements that are present in s8 but not in s9, which is {4, 5, 6, 7} because those are the elements that are in s8 but not in s9.
```

```python
s9 - s8
```

```python
s7 = {1,2,3,4,5} 
s8 = {4,5,6,7,8} 
s9 = {8,9,10} 
```

```python
s8.difference_update(s7)
# The difference_update() method updates the set by removing the elements that are present in another set. In this case, s8.difference_update(s7) will modify s8 to contain only the elements that are in s8 but not in s7, which will result in s8 being updated to {6, 7, 8}.
```

```python
s8
```

```python
s7 = {1,2,3,4,5} 
s8 = {4,5,6,7,8} 
s9 = {8,9,10} 
```

```python
s7.symmetric_difference(s8) 
# The symmetric_difference() method returns a new set that contains the elements that are in either set, but not in both. In this case, s7.symmetric_difference(s8) will return {1, 2, 3, 6, 7, 8} because those are the elements that are in either s7 or s8 but not in both.
```

```python
s8 ^ s9 
# The ^ operator is used to perform the symmetric difference of two sets. In this case, s8 ^ s9 will return a new set that contains the elements that are in either s8 or s9 but not in both, which is {4, 5, 6, 7, 9, 10} because those are the elements that are in either s8 or s9 but not in both.
```

```python
s7 = {1,2,3,4,5} 
s8 = {4,5,6,7,8} 
s9 = {8,9,10} 
```

```python
s7.symmetric_difference_update(s8) 
# The symmetric_difference_update() method updates the set with the symmetric difference of itself and another set. In this case, s7.symmetric_difference_update(s8) will modify s7 to contain only the elements that are in either s7 or s8 but not in both, which will result in s7 being updated to {1, 2, 3, 6, 7, 8}.
```

```python
a = {1,2,3,4,5,6,7,8,9} 
b = {3,4,5,6,7,8} 
c = {10,20,30,40} 
```

```python
a.issuperset(b)
# The issuperset() method returns True if the set is a superset of another set, meaning that it contains all the elements of the other set. In this case, a.issuperset(b) will return True because set a contains all the elements of set b.
```

```python
b.issuperset(a)
```

```python
b.issuperset(c)
```

```python
c.isdisjoint(a) 
# The isdisjoint() method returns True if the set has no elements in common with another set. In this case, c.isdisjoint(a) will return True because there are no common elements between set c and set a.
```

```python
a.isdisjoint(b)
```
