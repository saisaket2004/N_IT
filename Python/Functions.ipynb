```python
def myFun(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)

    print("\nKeyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

myFun('Hey', 'Welcome', first='Geeks', mid='for', last='Geeks')

#*args is used to pass a variable number of non-keyword arguments to a function, while **kwargs is used to pass a variable number of keyword arguments to a function.
```

```python
# In Python, *args and **kwargs are used to pass a variable number of arguments to a function.”

# *args allows us to pass multiple positional arguments, and it stores them as a tuple.
# **kwargs allows us to pass multiple keyword arguments, and it stores them as a dictionary.
```

```python
def test(*args):
    print(args)

test(1, 2, 3)
```

```python
def test(**kwargs):
    print(kwargs)

test(a=10, b=20)
```

```python
def test(*args, **kwargs):
    print(args)
    print(kwargs)

test(1, 2, x=100, y=200)
```

```python
#Keyword Arguments
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(name="Saket", age=21)
#student(age=21, name="Saket")
#Keyword Arguments allow you to specify the arguments by their names, making the code more readable and allowing you to pass arguments in any order.
```
