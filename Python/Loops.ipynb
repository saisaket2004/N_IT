```python
i = 1          # initializing

while i<=5:    # condition
    print('data science')
    i = i + 1  # increment
```

```python
i = 5          # initializing

while i>=1:    # condition
    print('data science :', i)
    i = i - 1  # increment
```

```python
i = 1

while i<=5:
    print('data science') # when we mention end then new line will not create
    j = 1
    while j<=4:
        print('technology')
        j = j + 1
        
    i = i + 1
    print()    
```

```python
i = 1
while i<=5:
    print(' datascience', end = "") # when we mention end then new line will not create
    j = 1
    while j<=4:
        print(' technology', end= "")
        j = j + 1
        
    i = i + 1
    print()
```

```python
i = 1

while i <= 2 :
    j = 0
    
    while  j <= 2 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```

```python
i = 1
while i <= 4 :
    j = 0    
    while  j <= 3 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1 
```
