```python
# There are 11 inbuilt functions in Python that can be used with lists. These functions are:
# 1. append(): Adds an element to the end of the list.
# 2. extend(): Adds all the elements of an iterable (e.g., another list) to the end of the list.
# 3. insert(): Inserts an element at a specified position in the list.  
# 4. remove(): Removes the first occurrence of a specified element from the list.
# 5. pop(): Removes and returns the element at a specified position in the list. If no position is specified, it removes and returns the last element.
# 6. clear(): Removes all elements from the list, resulting in an empty list.
# 7. index(): Returns the index of the first occurrence of a specified element in the list.
# 8. count(): Returns the number of occurrences of a specified element in the list.
# 9. sort(): Sorts the elements of the list in ascending order (or in a specified order).
# 10. reverse(): Reverses the order of the elements in the list.
# 11. copy(): Returns a shallow copy of the list.
         
```

```python
#There are two ways to create a copy of a list in Python: using the copy() method or using slicing. Both methods create a new list that is a copy of the original list, but they do not affect the original list when changes are made to the copied list.
l2 = lst.copy()
print(l2)
l3 = lst[:]
print(l3)
```

```python
lst.extend([5,6])
#The extend() method is used to add all the elements of an iterable (e.g., another list) to the end of the list. In this example, we are extending the lst list by adding the elements 5 and 6 to it.
lst
```

```python
lst.insert(1,1000)
#The insert() method is used to insert an element at a specified position in the list. In this example, we are inserting the value 1000 at index 1 of the lst list. This means that 1000 will be placed at the second position in the list (since indexing starts at 0), and the existing elements will be shifted to the right to accommodate the new element.
lst
```

```python
lst.remove(1000)
#The remove() method is used to remove the first occurrence of a specified element from the list.
lst
```
