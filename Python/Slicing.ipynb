# Slicing

```python
# slicing is a process of obtaining a substring from a string by specifying the start and end index.
# syntax: string[start:end:step]
```

```python
s1 = "welcome "
```

```python
s1[-3]
```

```python
print(s1[1:4]) #forward slicing
print(s1[-6:-3]) #backward slicing
```

```python
s1[:] # print all string
```

```python
s1[3:] # left index starting from 3rd till last
s1[:6] # from start index to (n-1) which is 5th idx
s1[2:9] # from 2nd idx to 8th idx
s1[2:9:2] # print from 2nd idx to 8th idx but you need to only print 2nd step
s1[:100] # if we give the end index more than the length of the string it will not give error it will print till the end of the string
s1[100:] # if we give the start index more than the length of the string it will not give error it will print empty string
```

```python
s1[::1] # print all string but you need to only print in 1st step
s1[::2] # print all string but you need to only print 2nd step
s1[::3] # print all string but you need to only print 3rd step
```

```python
print(s1[::-1]) # print all string but you need to only print in reverse order
print(s1[::-2]) # print all string but you need to only print in reverse order and 2nd step
print(s1[::-3]) # print all string but you need to only print in reverse order and 3rd step
```
