# Python Datatype

```python
i = 5 
i
type(i)
# the type of i is <class 'int'> because it is an integer.
```

```python
type()
# there is an error because the type() function requires an argument to determine the type of the object.
```

```python
id(i)
# the id() function returns the identity of an object, which is a unique integer that remains constant for the object during its lifetime. It can be used to check if two variables refer to the same object in memory.
```

```python
f = 110.45
f 
type(f)
# the type of f is <class 'float'> because it is a floating-point number.
```

```python
10 + 5
10 - 5
10 * 5
# only recent operation is shown, so the mul operation is only displayed here. to get all the output we need to use print() function.
```

```python
print(10+5)
print(10-5)
print(10*5)
```

```python
f1 = 2.4e1
f1
# the value of f1 is 24.0, which is a floating-point number in scientific notation. The "e" stands for exponent, and "2.4e1" means 2.4 multiplied by 10 to the power of 1, which equals 24.0.
```

```python
f2 = 3.5e3
f2
# the value of f2 is 3500.0, which is a floating-point number in scientific notation. The "e" stands for exponent, and "3.5e3" means 3.5 multiplied by 10 to the power of 3, which equals 3500.0.
```

```python
f3 = 3.2E3
f3
# the value of f3 is 3200.0, which is a floating-point number in scientific notation. The "E" stands for exponent, and "3.2E3" means 3.2 multiplied by 10 to the power of 3, which equals 3200.0.
```

```python
True
# True is a boolean value in Python that represents the truth value of an expression.
```

```python
False
# False is a boolean value in Python that represents the falsehood value of an expression.
```

```python
True + True
# the result of True + True is 2 because in Python, the boolean value True is treated as 1 and False is treated as 0 when used in arithmetic operations. Therefore, True + True is equivalent to 1 + 1, which equals 2.
```

```python
True + False - False 
# the result of True + False - False is 1 because True is treated as 1 and False is treated as 0 in arithmetic operations. Therefore, True + False - False is equivalent to 1 + 0 - 0, which equals 1.
```

```python
print(False / True)  #float division
print(False // True )
#
```

```python
c = 10 + 20j # 10 - real part & 20 imaginary part where j = squre root of -1 
print(type(c))
c 
# the value of c is 10 + 20j, which is a complex number in Python. The real part is 10 and the imaginary part is 20, where j represents the square root of -1.
```

```python
c.real
# this will return the real part of the complex number c, which is 10.0.
```

```python
c.imag
# this will return the imaginary part of the complex number c, which is 20.0.
```
