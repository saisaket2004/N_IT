```python
"""
Type Conversion in Python

1. Most data types can be converted to int or float,
   except complex numbers and text strings.

2. Most data types can be converted to bool.
   - Non-zero value → True
   - Zero value → False

3. Most data types can be converted to complex,
   but you cannot pass two strings together.

4. Any data type can be converted to string,
   except when it is already a text string.

5. Implicit Type Casting:
   Automatic conversion done by Python.

6. Explicit Type Casting:
   Manual conversion done by the programmer.
"""
```

```python
"""
Type Casting

• int / float → works for most types except complex and text strings  
• bool → non-zero = True, zero = False  
• complex → works for numbers (not two strings)  
• string → almost any type can convert to string  

Implicit Casting → automatic by Python  
Explicit Casting → manual by programmer

"""
```
