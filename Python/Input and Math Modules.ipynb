```python
math.floor(3.8)
# the floor function returns the largest integer less than or equal to a given number. In this case, math.floor(3.8) will return 3, because 3 is the largest integer that is less than or equal to 3.8.
```

```python
math.ceil(3.8)
# the ceil function returns the smallest integer greater than or equal to a given number. In this case, math.ceil(3.8) will return 4, because 4 is the smallest integer that is greater than or equal to 3.8.
```

```python
pow(3,2)
# the pow function takes two arguments and returns the result of raising the first argument to the power of the second argument. In this case, pow(3, 2) will return 9, because 3 raised to the power of 2 is equal to 3 * 3, which equals 9.
```
