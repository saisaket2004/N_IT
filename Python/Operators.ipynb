```python
name1 = "Alice" 
name2 = "Bob" 
print(name1 < name2)
# In Python, strings are compared lexicographically, which means that they are compared character by character based on their Unicode code points. In this case, "Alice" is considered less than "Bob" because the first character 'A' has a lower Unicode code point than 'B'. Therefore, the expression name1 < name2 evaluates to True.
```

```python
name1 = "Alice" 
name2 = "Bob" 
print(name1 > name2)
# In Python, strings are compared lexicographically, which means that they are compared character by character based on their Unicode code points. In this case, "Alice" is considered less than "Bob" because the first character 'A' has a lower Unicode code point than 'B'. Therefore, the expression name1 > name2 evaluates to False.
```

```python
name3 = 'priyanaka'
name4 = 'pc'
print(name3<=name4)
# In Python, strings are compared lexicographically, which means that they are compared character by character based on their Unicode code points. In this case, "priyanaka" is considered less than "pc" because the first character 'p' has the same Unicode code point in both strings, but the second character 'r' in "priyanaka" has a higher Unicode code point than 'c' in "pc". Therefore, the expression name3 <= name4 evaluates to False.
```
